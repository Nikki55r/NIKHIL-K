```python
d.drop('Serial No.',axis=1,inplace=True)
```

```python
d.head()
```

```python
d.hist(figsize=(14,6),bins=20);
```

```python
from scipy.stats import skew
dsk=d.apply(lambda x:skew(x.dropna()))
dsk=dsk[dsk>.75]

d[dsk.index]=np.log1p(d[dsk.index])
```

```python
d.hist(figsize=(14,6),bins=20);
```

```python
d=(d-d.min())/(d.max()-d.min())
d.describe()
```

```python
d.hist(figsize=(14,6));
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(d,t_lg,train_size=.7,random_state=1)
```

```python
import statsmodels.api as sm
mod=sm.OLS(y_train,x_train).fit()
```

```python
pred=mod.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(np.sqrt(mean_squared_error(pred,y_test)))
```
